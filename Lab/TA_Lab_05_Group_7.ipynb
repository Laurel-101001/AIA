Chart 1 — Survival Count by Gender
Show how many males and females survived.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

# 讀取鐵達尼號訓練資料
titanic_train = pd.read_csv('./titanic/train.csv')

sns.countplot(x='Sex', hue='Survived', data=titanic_train)
plt.title("1. Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.show()
# Description: This chart shows that a much higher proportion of females survived compared to males.
```

Chart 2 — Fare by Passenger Class
Compare the average fare among 1st, 2nd, and 3rd class passengers.

```python
sns.barplot(x='Pclass', y='Fare', data=titanic_train)
plt.title("2. Average Fare among 1st, 2nd, and 3rd Class Passengers")
plt.xlabel("Passenger Class (Pclass)")
plt.ylabel("Average Fare")
plt.show()
# Description: This chart demonstrates that 1st class passengers paid significantly higher fares on average.
```

Chart 3 — Age Distribution of Survivors
Show how the ages of survivors are distributed.

Hint: Survived, Age columns are needed

```python
import seaborn as sns
import matplotlib.pyplot as plt
survivors = titanic_train[titanic_train['Survived'] == 1]

sns.histplot(survivors['Age'], bins = 20, kde = True)
plt.title("3. Age Distribution of Survivors")
plt.xlabel("Age")
plt.ylabel("Number of Survivors")
plt.show()
# Description: This chart reveals that most survivors were young adults in their 20s and 30s, along with some infants.
```
